# Decision Tree Classification

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

### Step 1 - Read the data

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df
```

```python
df["species"].value_counts()
```

### Step 2 -  Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates(keep="first")
```

```python
df.shape
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.index
```

### Step 3 - Separate X and Y(species)

```python
x= df.drop(columns=["species"])
y=df[["species"]]
```

```python
x.head()
```

```python
y.head()
```

### Apply preprocessing on x

```python
x.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
x_pre = num_pipe.fit_transform(x)
x_pre.head()
```

```python
num_pipe
```

### Step 5 - Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(
    x_pre, y, test_size =0.33, random_state =21
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 - Model Building

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="entropy"
)
dtc.fit(xtrain, ytrain)
```

```python
dtc.score(xtrain,ytrain)
```

```python
dtc.score(xtest, ytest)
```

### Hyperparameter tuning to improve model performance

```python
params={
    "max_depth": [2,3,4,5,6],
    "min_samples_split":[5,6,7,8,9],
    "min_samples_leaf":[5,6,7,8,9],
    "criterion":["gini", "entropy"]
}
```

```python
from sklearn.model_selection import GridSearchCV
model=DecisionTreeClassifier()
gscv =GridSearchCV(
    estimator=model,
    param_grid=params,
    scoring="f1_macro",
    cv=5
)
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Step 6- Model evaluation

```python
best_dtc.score(xtrain,ytrain)
```

```python
best_dtc.score(xtest,ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc,xtest,ytest)
```

```python
ypred_test = best_dtc.predict(xtest)
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
